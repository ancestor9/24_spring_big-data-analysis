### **2. 앨범 발매 년도별  노래 건수**

```python
# prompt: release_date를 날짜데이터로 변환
pd.to_datetime(df['release_date'])
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
df.head()
```

```python
df['release_date'].dt.year
```

```python
df['release_date'].dt.month
```

```python
df['년도'] = df['release_date'].dt.year
df['월'] = df['release_date'].dt.month
```

```python
df.head()
```

```python
df.head()
```

```python
df['release_date'].value_counts()
```

```python
# 앨범발매연도별노래개수의합
# 앨범발매연월별노래개수의합
```

```python
df.groupby(['album_title', '년도']).size().reset_index(name='song_count')
```

```python
df.head()
```

```python
df.info()
```

```python
# prompt: 년도와 월을 합치는 코드

df['발매년월'] = df['년도'].astype(str) + '-' + df['월'].astype(str)
df.head()
```

```python
df.groupby(['album_title', '발매년월']).size().reset_index(name='song_count')
```
