제대로 읽는 방법

```python
# ZIP 파일 내의 Excel 파일 리스트 얻기
excel_files = [name for name in zipfile.namelist() if name.endswith('.xlsx')]
excel_files
```

```python
# ZIP 파일 내의 Excel 파일 리스트 얻기
excel_files = [name for name in zipfile.namelist() if name.endswith('.xlsx')]

# 각 Excel 파일을 읽어서 데이터프레임으로 저장
dfs = []
for excel_file in excel_files:
    with zipfile.open(excel_file) as file:
        # Excel 파일을 데이터프레임으로 읽기
        df = pd.read_excel(file)
        dfs.append(df)

# 결과 데이터프레임 리스트를 확인하거나 합칠 수 있음
for idx, df in enumerate(dfs):
    print(f"DataFrame {idx + 1}:\n{df}\n")
```
