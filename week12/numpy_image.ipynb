```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Read the image
image = cv2.imread('brad.jpg')

# Convert the image from BGR to RGB color space
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Define the sheer transformation matrix
shear_matrix = np.float32([[1, 0.5, 0], [0, 1, 0]])

# Apply the sheer transformation to the image
image_sheared = cv2.warpAffine(image_rgb, shear_matrix, (image_rgb.shape[1], image_rgb.shape[0]))

# Display the original and sheared images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(image_sheared)
plt.title('Sheared Image')

plt.show()

```
