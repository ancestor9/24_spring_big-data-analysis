## **Build My Own Module(functions)**

```python
%%writefile utilities.py


def add(x, y):

    return x + y
def subtract(x, y):
    return x - y
```

```python
print(__name__)
```

```python
from utilities import add, subtract

def main():
    print("Adding 1 + 2 gives:", add(1, 2))
    print("Subtracting 5 - 3 gives:", subtract(5, 3))

if __name__ == "__main__":
    main()
```
