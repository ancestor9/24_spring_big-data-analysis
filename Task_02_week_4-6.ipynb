### **('구분자').join(iterables) vs. split('text')**

```python
' '.join(df['song_lyrics'])
```

```python
import re
# 'song_lyrics' 컬럼의 모든 문장을 하나의 문자열로 합치기
merged_lyrics = ' '.join(df['song_lyrics'])

# 정규 표현식을 사용하여 불필요한 문자 및 숫자 제거
cleaned_lyrics = re.sub(r'[^a-zA-Z\s]', '', merged_lyrics)
cleaned_lyrics
```

>> ### **chat-gpt : 한글도 포함하려면**

```python
# 정규 표현식을 사용하여 알파벳, 공백 문자, 한글 이외의 모든 문자 및 숫자를 제거
cleaned_text = re.sub(r'[^\w\sㄱ-ㅎㅏ-ㅣ가-힣]', '', merged_lyrics)
cleaned_text
```

>> ### **chat-gpt : 연속된 공백 문자를 한 칸만 남기고 대체**

```python
cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()
cleaned_text
```

```python
cleaned_text[:20].split()
```

```python
Counter(cleaned_text[:20].split())
```

>> ### **chat-gpt :cleaned_text에서 단어의 출현빈도를 알고싶다.**

```python
from collections import Counter

# 문자열을 공백을 기준으로 단어로 분할하고 단어 빈도 계산
word_list = cleaned_text.split()
word_freq = Counter(word_list)

# 단어 빈도 출력
print(word_freq)
```

```python
word_freq.most_common()[:10]
```

```python
# 단어 빈도를 높은 빈도 순으로 정렬하여 출력
sorted_word_freq = word_freq.most_common()
for word, freq in sorted_word_freq:
    print(f'{word}: {freq}')
```
