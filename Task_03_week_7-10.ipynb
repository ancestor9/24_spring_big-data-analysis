>> ###### KNU 감성어 사전은 json 파일로 만들어져 있음으로 json 라이브러리를  호출해 사전 파일 SentiWord_info.json을 열고, 영화리뷰와 비교할 수 있도록 데이터프레임으로 만들도록 하자.
- 감성어 사전 링크:

https://github.com/park1200656/KnuSentiLex/tree/master/KnuSentiLex/data

```python
import pandas as pd
import requests

# Step 1: Use the raw URL of the JSON file
url = 'https://raw.githubusercontent.com/park1200656/KnuSentiLex/master/KnuSentiLex/data/SentiWord_info.json'

# Step 2: Fetch the content
response = requests.get(url)

if response.status_code == 200:
    # Step 3: Save the content to a file and read using pandas
    with open('SentiWord_info.json', 'w') as f:
        f.write(response.text)

    # Step 4: Load the JSON into a pandas DataFrame
    sentiword_dic = pd.read_json('SentiWord_info.json')
    display(sentiword_dic.sample(10))
else:
    print("Failed to retrieve content from the URL.")
```

> ## **3. 토큰화된 문장과 감성사전 비교**
- 이제 온라인 기사 한 문장씩을 넣어서 해당 문장에 있는 단어와 감성사전 단어를 매치시켜 존재하면 극성값을 가져오고, 없으면 패스시키도록 하자.
- 이때 문장 하나를 가져와 단어를 하나씩 끄집어 넣어 감성 사전 단어와 비교해야 하기 때문에 2중  for 문을 이용해야 한다.
- 리뷰 극성값은 문장의 포함된 단어의 극성값 합계로 계산한다. 예를 들어, "나는 이 영화가 재밌고 감동적이다"라면, '재밌고'가 +1, 감동적이다 +2점으로 이 문장은 긍정적인 +3점 리뷰로 판단된다.

```python
tf = pd.DataFrame(columns=("review", "sentiment"))  # 리뷰별 극성을 저장하기 위한 데이터프레임 생성
tf
```

```python
from tqdm import tqdm
```

```python
from time import time       # 아래 코드가 몇시간 걸리나
time()          # 현재 시간을 알려줌
```

```python
start = time() # 시작시간 기록

idx = 0                                             # 다음 리뷰로 넘기기 위한 초기값

for token in tqdm(tokens):                                # 전체 리뷰에서 문장 하나씩 가져옴
  sentiment = 0                                     # 초기 감성값 0으로 설정
  for i in range(0, len(sentiword_dic)):            # 감성사전의 모든 단어를 하나씩 선택
    if sentiword_dic.word[i] in token:              # 리뷰 문장에 감성 단어가 있는지 확인
      sentiment += int(sentiword_dic.polarity[i])   # 감성단어가 있다면 극성값 합계를 구함.
  tf.loc[idx] = [token, sentiment]                  # 리뷰별 극성값을 데이터프레임으로 쌓음
  idx += 1                                          # 다름 리뷰 문장으로 넘어감

end = time()  # 끝나는 시간 기록

print(f'time elapsed: {end - start}')
```
